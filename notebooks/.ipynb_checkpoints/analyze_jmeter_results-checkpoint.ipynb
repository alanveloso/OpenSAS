# Análise Estatística dos Resultados JMeter por Tipo de Requisição

Este notebook lê os arquivos `.jtl` organizados por tipo de requisição (ex: Authorize, Registration, etc.) e realiza análises estatísticas considerando múltiplos runs.

- Estatísticas: média, mediana, desvio padrão, percentis, taxa de sucesso, etc.
- Visualizações: boxplot, histogramas, linhas de tendência.

> **Pré-requisito:** Os resultados devem estar organizados em subpastas por tipo de requisição dentro de cada cenário em `results/`.


```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# pasta onde estão os resultados
RESULTS_DIR = '../results'  # ajuste se necessário

# níveis e ordem fixa desejada
desired_levels = ['low', 'medium', 'high', 'stress']

# carrega todos os .jtl de um dado request_type
def load_all_by_type(request_type):
    pattern = os.path.join(RESULTS_DIR, '*', request_type, '*.jtl')
    files = glob.glob(pattern)
    dfs = []
    for f in files:
        df = pd.read_csv(f)
        df['run_file'] = os.path.basename(f)
        df['scenario'] = os.path.basename(os.path.dirname(os.path.dirname(f)))
        dfs.append(df)
    return pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()

# descobre todos os tipos de requisição disponíveis em RESULTS_DIR
def get_request_types():
    types = set()
    for scenario in os.listdir(RESULTS_DIR):
        sp = os.path.join(RESULTS_DIR, scenario)
        if os.path.isdir(sp):
            for req in os.listdir(sp):
                if os.path.isdir(os.path.join(sp, req)):
                    types.add(req)
    return sorted(types)

# lista de tipos e print inicial
request_types = get_request_types()

print('Tipos de requisição encontrados:', request_types)
```

## Estatísticas por Tipo de Requisição

Para cada tipo, serão exibidas estatísticas e gráficos.


```python
for req_type in request_types:
    print(f'\n--- {req_type} ---')
    df = load_all_by_type(req_type)
    if df.empty:
        print('No data.')
        continue

    # Basic statistics
    stats = df['elapsed'].describe(percentiles=[.5, .9, .95, .99])
    success_rate = df['success'].mean() * 100
    print(stats)
    print(f'Success rate: {success_rate:.2f}%')

    # Extract only the level suffix and capitalize it
    df['level'] = (
        df['scenario']
          .str.split('_')
          .str[-1]
          .str.title()
    )

    # Ensure the correct ordering of levels
    level_labels = [lvl.title() for lvl in desired_levels]
    df['level'] = pd.Categorical(df['level'],
                                 categories=level_labels,
                                 ordered=True)

    # Boxplot using 'level' instead of full scenario
    plt.figure(figsize=(10, 4))
    ax = sns.boxplot(x='level',
                     y='elapsed',
                     data=df,
                     order=level_labels)
    ax.set_title(f'Response Time Boxplot – {req_type}')
    ax.set_xlabel('Level')
    ax.set_ylabel('Elapsed (ms)')
    # labels sem rotação
    ax.tick_params(axis='x', rotation=0)
    plt.tight_layout()
    plt.show()

    # Histogram of elapsed times
    plt.figure(figsize=(8, 3))
    sns.histplot(df['elapsed'], bins=30, kde=True)
    plt.title(f'Response Time Histogram – {req_type}')
    plt.xlabel('Elapsed (ms)')
    plt.tight_layout()
    plt.show()
```
